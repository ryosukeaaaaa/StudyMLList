```python
import matplotlib.pyplot as plt
import numpy as np

# シグモイド巻数を定義
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# 0.1間隔で-7から7までの値を生成
z = np.arange(-7, 7, 0.1)
# シグモイド関数を計算
sigma_z = sigmoid(z)

plt.plot(z, sigma_z)
# 垂直線を追加(z=0)
plt.axvline(0.0, color='k')
# y軸の範囲を設定
plt.ylim(-0.1, 1.1)
# 軸のラベルを設定
plt.xlabel('z')
plt.ylabel(r'$\sigma (z)$')
# y軸の目盛りを設定
plt.yticks([0.5*i for i in range(3)])
# Avesクラスのオブジェクトを取得
ax = plt.gca()
# グリッドを追加
ax.yaxis.grid(True)
# グラフを表示
plt.tight_layout()
plt.show()
```
